```python
!pip install scikit-fuzzy
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

import skfuzzy as fuzz
```

```python

# 1) Define variables por tipo (alineado a objetivos)
vars_demo = ["EDAD", "SEXO", "ORIGEN"]
vars_comorb = ["DIABETES", "HIPERTENSION", "OBESIDAD", "TABAQUISMO"]
vars_result = ["TIPO_PACIENTE", "NEUMONIA", "INTUBADO", "UCI", "DEFUNCION"]

# 2) Variables para clustering (recomendación: demográficas + comorbilidades)
vars_cluster = vars_demo + vars_comorb

# 3) Asegurar que existan en ambos años
comun = set(datos_covid_21.columns).intersection(set(datos_covid_23.columns))
vars_demo = [c for c in vars_demo if c in comun]
vars_comorb = [c for c in vars_comorb if c in comun]
vars_result = [c for c in vars_result if c in comun]
vars_cluster = vars_demo + vars_comorb

print("Variables demográficas:", vars_demo)
print("Comorbilidades:", vars_comorb)
print("Resultados:", vars_result)
print("Variables clustering:", vars_cluster)

df21 = datos_covid_21[vars_cluster + vars_result].copy()
df23 = datos_covid_23[vars_cluster + vars_result].copy()
```

```python
# Subconjuntos de trabajo (ya con EDAD acotada)
df21 = datos_covid_21[vars_cluster + vars_result].copy()
df23 = datos_covid_23[vars_cluster + vars_result].copy()
```

```python
import numpy as np
import pandas as pd

def recodificar_si_no(df, cols):
    df = df.copy()
    for c in cols:
        if c in df.columns:
            df[c] = df[c].where(df[c].isin([1,2]), np.nan)
            df[c] = df[c].map({1:1, 2:0})
    return df

def recodificar_sexo_origen(df):
    df = df.copy()
    if "SEXO" in df.columns:
        df["SEXO"] = df["SEXO"].where(df["SEXO"].isin([1,2]), np.nan)
        df["SEXO"] = df["SEXO"].map({1:"MUJER", 2:"HOMBRE"})
    if "ORIGEN" in df.columns:
        df["ORIGEN"] = df["ORIGEN"].where(df["ORIGEN"].isin([1,2]), np.nan)
        df["ORIGEN"] = df["ORIGEN"].map({1:"USMER", 2:"NO_USMER"})
    return df

# aplicar
bin_cols = [c for c in vars_comorb + ["NEUMONIA","INTUBADO","UCI"] if c in df21.columns]
df21 = recodificar_si_no(df21, bin_cols)
df23 = recodificar_si_no(df23, bin_cols)

df21 = recodificar_sexo_origen(df21)
df23 = recodificar_sexo_origen(df23)
```

```python
from sklearn.preprocessing import StandardScaler

def construir_X(df, cols):
    X = df[cols].copy()
    X = pd.get_dummies(X, drop_first=False)
    for c in X.columns:
        if pd.api.types.is_numeric_dtype(X[c]):
            X[c] = X[c].fillna(X[c].median())
    Xs = StandardScaler().fit_transform(X)
    return pd.DataFrame(Xs, columns=X.columns, index=X.index)

# Escenario A: SOLO DEMOGRÁFICO
X21_demo = construir_X(df21, vars_demo)
X23_demo = construir_X(df23, vars_demo)
X21_demo, X23_demo = X21_demo.align(X23_demo, join="outer", axis=1, fill_value=0)

# Escenario B: DEMO + CLÍNICO (principal)
X21 = construir_X(df21, vars_cluster)
X23 = construir_X(df23, vars_cluster)
X21, X23 = X21.align(X23, join="outer", axis=1, fill_value=0)

X21_demo.shape, X23_demo.shape, X21.shape, X23.shape
```

* One-Hot para categorias. Evitar el orden falso.

* Escalar evitará que cierta edad domine.

* Alineamos las columnas para garantizar la comparabilidad entre los años.

```python
from sklearn.cluster import KMeans

def kmeans_k2_muestreo(X, n_muestra=200_000, seed=42):
    Xm = X.sample(n=min(n_muestra, len(X)), random_state=seed)
    km = KMeans(n_clusters=2, n_init=10, random_state=seed).fit(Xm)
    labels = km.predict(X)
    return km, labels

# Demográfico
_, lab21_demo = kmeans_k2_muestreo(X21_demo)
_, lab23_demo = kmeans_k2_muestreo(X23_demo)
df21["cluster_kmeans_demo"] = lab21_demo
df23["cluster_kmeans_demo"] = lab23_demo

# Demo + Clínico
_, lab21_dc = kmeans_k2_muestreo(X21)
_, lab23_dc = kmeans_k2_muestreo(X23)
df21["cluster_kmeans_demo_clin"] = lab21_dc
df23["cluster_kmeans_demo_clin"] = lab23_dc

df21["cluster_kmeans_demo"].value_counts(), df23["cluster_kmeans_demo"].value_counts()
```

*  Debido al gran tamaño del conjunto de datos, la caracterización de los clústeres se realizó sobre una muestra aleatoria representativa, lo que permite reducir la carga computacional sin perder la estructura general de los datos.

* k=2, ya que preferimos asignar clusters de acuerdo a la edad codificada.

```python
def muestra_estratificada(df, col_cluster, n_muestra=200_000, seed=42):
    n_total = len(df)
    return (df.groupby(col_cluster, group_keys=False)
              .apply(lambda x: x.sample(
                  n=min(len(x), max(1, int(n_muestra * len(x) / n_total))),
                  random_state=seed)))

df21_demo_m = muestra_estratificada(df21, "cluster_kmeans_demo")
df23_demo_m = muestra_estratificada(df23, "cluster_kmeans_demo")

df21_dc_m = muestra_estratificada(df21, "cluster_kmeans_demo_clin")
df23_dc_m = muestra_estratificada(df23, "cluster_kmeans_demo_clin")
```

Esto nos permite mantener las proporciones de los clusteres.

```python
# --- 7A) Perfil DEMO + CLÍNICO: numéricas/binarizadas (mean = proporciones) ---
cols_num = []

# EDAD (numérica)
if "EDAD" in df21.columns:
    cols_num.append("EDAD")

# Comorbilidades (0/1)
cols_num += [c for c in vars_comorb if c in df21.columns]

print("Perfil numérico (EDAD + comorbilidades) por clúster – 2021 (muestra)")
display(df21_dc_m.groupby("cluster_kmeans_demo_clin")[cols_num].mean().round(3))

print("Perfil numérico (EDAD + comorbilidades) por clúster – 2023 (muestra)")
display(df23_dc_m.groupby("cluster_kmeans_demo_clin")[cols_num].mean().round(3))


# --- 7B) Perfil DEMO + CLÍNICO: categóricas (SEXO / ORIGEN) como proporciones ---
if "SEXO" in df21.columns:
    print("SEXO por clúster – 2021 (muestra)")
    display(pd.crosstab(df21_dc_m["cluster_kmeans_demo_clin"], df21_dc_m["SEXO"], normalize="index").round(3))

    print("SEXO por clúster – 2023 (muestra)")
    display(pd.crosstab(df23_dc_m["cluster_kmeans_demo_clin"], df23_dc_m["SEXO"], normalize="index").round(3))

if "ORIGEN" in df21.columns:
    print("ORIGEN por clúster – 2021 (muestra)")
    display(pd.crosstab(df21_dc_m["cluster_kmeans_demo_clin"], df21_dc_m["ORIGEN"], normalize="index").round(3))

    print("ORIGEN por clúster – 2023 (muestra)")
    display(pd.crosstab(df23_dc_m["cluster_kmeans_demo_clin"], df23_dc_m["ORIGEN"], normalize="index").round(3))


# --- 7C) Relación clúster vs resultados (solo columnas numéricas/binarizadas) ---
cols_res = [c for c in ["NEUMONIA","INTUBADO","UCI","DEFUNCION"] if c in df21.columns]

# Si TIPO_PACIENTE es 1/2, recodifícalo antes; si ya es numérico 0/1, entonces inclúyelo:
if "TIPO_PACIENTE" in df21.columns and pd.api.types.is_numeric_dtype(df21["TIPO_PACIENTE"]):
    cols_res.append("TIPO_PACIENTE")

print("Relación clúster vs resultados – 2021 (muestra)")
display(df21_dc_m.groupby("cluster_kmeans_demo_clin")[cols_res].mean().round(3))

print("Relación clúster vs resultados – 2023 (muestra)")
display(df23_dc_m.groupby("cluster_kmeans_demo_clin")[cols_res].mean().round(3))
```

Los clusteres obtenidos presentan una separación clara por sexo, lo que indica que la variable tiene un peso relevante en la segmentación de los pacientes.


```python
import skfuzzy as fuzz

def fcm_entrenar_muestra(X, n_muestra=100_000, seed=42):
    Xm = X.sample(n=min(n_muestra, len(X)), random_state=seed)
    cntr, u, *_ , fpc = fuzz.cluster.cmeans(
        Xm.to_numpy().T, c=2, m=2.0, error=0.005, maxiter=300, seed=seed
    )
    return cntr, fpc

def asignar_por_centroides(X, cntr):
    Xn = X.to_numpy()
    d0 = np.linalg.norm(Xn - cntr[0], axis=1)
    d1 = np.linalg.norm(Xn - cntr[1], axis=1)
    return (d1 < d0).astype(int)

cntr21, fpc21 = fcm_entrenar_muestra(X21)
cntr23, fpc23 = fcm_entrenar_muestra(X23)

df21["cluster_fcm_demo_clin"] = asignar_por_centroides(X21, cntr21)
df23["cluster_fcm_demo_clin"] = asignar_por_centroides(X23, cntr23)

fpc21, fpc23
```

En el 2023 la partición fue un poco más difusa (0.61)

```python
# Variables numéricas/binarizadas
vars_tabla = (
    ["EDAD"] +
    [c for c in vars_comorb if c in df21.columns] +
    [c for c in ["NEUMONIA","INTUBADO","UCI","DEFUNCION"] if c in df21.columns]
)

def tabla_comparativa(df21, df23, col_cluster, vars_tabla):
    t21 = (
        df21
        .groupby(col_cluster)[vars_tabla]
        .mean()
        .rename(columns=lambda x: f"{x}_2021")
    )

    t23 = (
        df23
        .groupby(col_cluster)[vars_tabla]
        .mean()
        .rename(columns=lambda x: f"{x}_2023")
    )

    tabla = pd.concat([t21, t23], axis=1)
    return tabla.round(3)

tabla_comp = tabla_comparativa(
    df21_cmp,
    df23_cmp,
    "cluster_kmeans_demo_clin",
    vars_tabla
)

tabla_comp
```

La diferencia no es extrema, pero es coherente con lo que se dijo previamente.



* COMPARACIÓN TEMPORAL

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns



# Muestra para visualización
X21_plot = X21.sample(n=50_000, random_state=42)
labels21_plot = df21.loc[X21_plot.index, "cluster_kmeans_demo_clin"]

# PCA
pca21 = PCA(n_components=2, random_state=42)
X21_pca = pca21.fit_transform(X21_plot)

df21_pca = pd.DataFrame(X21_pca, columns=["PC1", "PC2"])
df21_pca["cluster"] = labels21_plot.values

# Scatter
plt.figure(figsize=(6,5))
sns.scatterplot(
    data=df21_pca,
    x="PC1",
    y="PC2",
    hue="cluster",
    alpha=0.5,
    s=20
)
plt.title("Visualización de clústeres (PCA) – 2021")
plt.xlabel("Componente principal 1")
plt.ylabel("Componente principal 2")
plt.legend(title="Clúster")
plt.tight_layout()
plt.show()

print("Varianza explicada PCA 2021:", pca21.explained_variance_ratio_.round(3))
```

Se observa una separación clara entre los clústeres principalmente a lo largo de la segunda componente principal. La presencia de patrones en bandas horizontales se debe a la naturaleza categórica y binaria de varias variables clínicas y demográficas incluidas en el análisis. El solapamiento parcial entre grupos refleja la existencia de pacientes con características intermedias, lo cual es consistente con los resultados obtenidos mediante Fuzzy C-means.
